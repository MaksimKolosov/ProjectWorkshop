```python
# Импорты
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Загружаем первоначальные данные
initial_data = pd.read_csv('files/train.csv')

subset1_sum = initial_data.loc[((initial_data['Tool wear [min]']>=0) & 
                            (initial_data['Tool wear [min]']<=20)),'Machine failure'].sum()
print(f"Износ 0 - 20 мин: {subset1_sum} отказов")

subset2_sum = initial_data.loc[((initial_data['Tool wear [min]']>20) & 
                            (initial_data['Tool wear [min]']<=40)),'Machine failure'].sum()
print(f"Износ 20 - 40 мин: {subset2_sum} отказов")

subset3_sum = initial_data.loc[((initial_data['Tool wear [min]']>40) & 
                            (initial_data['Tool wear [min]']<=60)),'Machine failure'].sum()
print(f"Износ 40 - 60 мин: {subset3_sum} отказов")

subset4_sum = initial_data.loc[((initial_data['Tool wear [min]']>60) & 
                            (initial_data['Tool wear [min]']<=80)),'Machine failure'].sum()
print(f"Износ 60 - 80 мин: {subset4_sum} отказов")

subset5_sum = initial_data.loc[((initial_data['Tool wear [min]']>80) & 
                            (initial_data['Tool wear [min]']<=100)),'Machine failure'].sum()
print(f"Износ 80 - 100 мин: {subset5_sum} отказов")

subset6_sum = initial_data.loc[((initial_data['Tool wear [min]']>100) & 
                            (initial_data['Tool wear [min]']<=120)),'Machine failure'].sum()
print(f"Износ 100 - 120 мин: {subset6_sum} отказов")

subset7_sum = initial_data.loc[((initial_data['Tool wear [min]']>120) & 
                            (initial_data['Tool wear [min]']<=140)),'Machine failure'].sum()
print(f"Износ 120 - 140 мин: {subset7_sum} отказов")

subset8_sum = initial_data.loc[((initial_data['Tool wear [min]']>140) & 
                            (initial_data['Tool wear [min]']<=160)),'Machine failure'].sum()
print(f"Износ 140 - 160 мин: {subset8_sum} отказов")

subset9_sum = initial_data.loc[((initial_data['Tool wear [min]']>160) & 
                            (initial_data['Tool wear [min]']<=180)),'Machine failure'].sum()
print(f"Износ 160 - 180 мин: {subset9_sum} отказов")

subset10_sum = initial_data.loc[((initial_data['Tool wear [min]']>180) & 
                            (initial_data['Tool wear [min]']<=200)),'Machine failure'].sum()
print(f"Износ 180 - 200 мин: {subset10_sum} отказов")

subset11_sum = initial_data.loc[((initial_data['Tool wear [min]']>200) & 
                            (initial_data['Tool wear [min]']<=220)),'Machine failure'].sum()
print(f"Износ 200 - 220 мин: {subset11_sum} отказов")

subset12_sum = initial_data.loc[((initial_data['Tool wear [min]']>220) & 
                            (initial_data['Tool wear [min]']<=240)),'Machine failure'].sum()
print(f"Износ 220 - 240 мин: {subset12_sum} отказов")

plt.plot(np.array([10, 30, 50, 70, 90, 110, 130, 150, 170, 190, 210, 230]), 
        np.array([265, 107, 91, 133, 138, 170, 102, 138, 186, 267, 426, 117]))

plt.xlabel("Количество минут износа")
plt.ylabel("Количество отказов")
```
